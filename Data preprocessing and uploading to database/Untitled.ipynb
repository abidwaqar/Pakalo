```python
# Loading data
recipes_json = pd.read_json("recipes_raw_nosource_epi.json")
```

```python
# Preprocessig
recipes_ndarray = recipes_json.T[['title', 'ingredients', 'instructions']].values
```

```python
recipes_ndarray[0][2]
```

```python
# Formating instructions
for i in range(recipes_ndarray.shape[0]):
    instructions = recipes_ndarray[i][2]
    if type(instructions) == float:
        continue
    str_list = instructions.replace('\n', '').replace('(', '').replace(')', '').split('.')
    str_list = list(filter(None, str_list))
    recipes_ndarray[i][2] = [steps.strip() for steps in str_list]
#     print(i)
    if i%1000 == 0:
        print (i)
        print (recipes_ndarray[i][2])
        print ()
```

```python
# Cleaning ingredients from ADVERTISEMENT Keyword
for i in range(recipes_ndarray.shape[0]):
    if type(recipes_ndarray[i][1]) == float:
        continue
    recipes_ndarray[i][1] = [ingredients.replace('ADVERTISEMENT', '') for ingredients in recipes_ndarray[i][1]]
    recipes_ndarray[i][1] = list(filter(None, recipes_ndarray[i][1]))
    recipes_ndarray[i][1] = [steps.strip() for steps in recipes_ndarray[i][1]]
```

```python
for i in range(recipes_ndarray.shape[0]):
    recipes_ndarray[i][0] = recipes_ndarray[i][0].strip()
```
